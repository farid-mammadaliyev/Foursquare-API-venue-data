# Determine optimal K for K-means Clustering using Elbow method

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python

istanbul_venues_grouped = pd.read_csv('istanbul_venues_grouped.csv')
```

```python
istanbul_venues_grouped.head(15)
```

# Define sum of distances for each K values in a range of 1-10

```python
del istanbul_venues_grouped['Neighborhood']
```

```python
Sum_of_squared_distances = []

K = range(1,10)
for k in K:
    km = KMeans(n_clusters = k)
    km = km.fit(istanbul_venues_grouped)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
Sum_of_squared_distances
```

# Plot sum of distances and determine optimal K using elbow

```python
figure = plt.figure(figsize = (12, 8))
plt.plot(K, Sum_of_squared_distances, 'k--')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```
